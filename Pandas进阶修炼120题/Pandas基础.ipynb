```python
#导入numpy和Pandas库
import numpy as np
import pandas as pd
```

```python
#创建字典
data = {"grammer":["Python","C","Java","GO",np.nan,"SQL","PHP","Python"],
        "score":[1,2,np.nan,4,5,6,7,10]}
data
```

```python
#创建DataFrame
df = pd.DataFrame(data=data)
df
```

```python
#搜索含有字符串Python的行
df[df["grammer"]=="Python"]
```

```python
#输出所有的列名
df.columns
```

```python
#修改第二列列名为'popularity'
df.rename(columns={'score':'popularity'},inplace=True)
df
```

```python
#统计grammer列中每种编程语言出现的次数
df['grammer'].value_counts()
```

```python
#将空值用上下值的平均值填充
df['populations'] = df['popularity'].fillna(df['popularity'].interpolate())
df
```

```python
#提取popularity列中值大于3的行
df[df['popularity']>3]
```

```python
#按照grammer列进行去除重复值
df.drop_duplicates(['grammer'])
```

```python
#计算popularity平均值
df['popularity'].mean()
```

```python
#将grammer列转换为list
df['grammer'].to_list()
```

```python
#将DataFrame保存为EXCEL
df.to_excel('test.xlsx')
```

```python
#查看数据行数
df.shape
```

```python
#提取popularity列值大于3小于7的行
df[(df['popularity']>3)&(df['popularity']<7)]
```

```python
#查看最后5行数据
df.tail()
```

```python
#删除最后一行数据
df.drop([len(df)-1],inplace=True)
df
```

```python
#添加一行数据['Perl',6.6]
row={'grammer':'Perl','popularity':6.6}
df = df.append(row,ignore_index=True)
df
```

```python
#对数据按照“popularity”列值的大小进行排序
df.sort_values("popularity",inplace=True)
df
```

```python
#统计grammer列每个字符串的长度
df['grammer']=df['grammer'].fillna('R')
df['len_str']=df['grammer'].map(lambda x:len(x))
df
```
