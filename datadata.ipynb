```python
from pyecharts.charts import *
from pyecharts import options as opts
from pyecharts.faker import Faker
```

```python
import pandas as pd
df = pd.read_excel(r'D:/kaggle/first.xlsx')
def bar_stack():
    bar = Bar(init_opts=opts.InitOpts(theme='light',width='1000px',height='600px'))
    bar.add_xaxis(list(df['Unnamed: 0']))
    bar.add_yaxis('国内生产总值',list(df['国内生产总值']),stack='stack1')
    bar.add_yaxis('出口',list(df['出口']),stack='stack1')
    bar.add_yaxis('进口',list(df['进口']),stack='stack1')
    bar.add_yaxis('国内生产总值',list(df['Unnamed: 2']),stack='stack2')
    bar.add_yaxis('出口',list(df['Unnamed: 4']),stack='stack2')
    bar.add_yaxis('进口',list(df['Unnamed: 6']),stack='stack2')
    bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    return bar
chart = bar_stack()
chart.render_notebook()
```

```python
import pandas as pd
df = pd.read_excel(r"D:/kaggle/second.xlsx")
df = df.drop(0)
def bar_stack():
    bar = Bar(init_opts=opts.InitOpts(theme='light',width='1500px',height='700px'))
    bar.add_xaxis(list(df['Unnamed: 0']))
    bar.add_yaxis('中国',list(df['中国']),stack='stack1',itemstyle_opts=opts.ItemStyleOpts(color='red',opacity=0.6))
    bar.add_yaxis('中国',list(df['Unnamed: 2']),stack='stack2',color='red',itemstyle_opts=opts.ItemStyleOpts(color='red',opacity=0.6))
    bar.add_yaxis('美国',list(df['美国']),stack='stack1',itemstyle_opts=opts.ItemStyleOpts(color='blue',opacity=0.6))
    bar.add_yaxis('美国',list(df['Unnamed: 4']),stack='stack2',itemstyle_opts=opts.ItemStyleOpts(color='blue',opacity=0.6))
    bar.add_yaxis('世界其他',list(df['世界其他']),stack='stack1',itemstyle_opts=opts.ItemStyleOpts(color='yellow',opacity=0.6))
    bar.add_yaxis('世界其他',list(df['Unnamed: 6']),stack='stack2',itemstyle_opts=opts.ItemStyleOpts(color='yellow',opacity=0.6))
    bar.add_yaxis('世界',list(df['世界']),stack='stack1',itemstyle_opts=opts.ItemStyleOpts(color='green',opacity=0.6))
    bar.add_yaxis('世界',list(df['Unnamed: 8']),stack='stack2',itemstyle_opts=opts.ItemStyleOpts(color='green',opacity=0.6))
    bar.set_series_opts(label_opts=opts.LabelOpts(False))
    return bar
chart = bar_stack()
chart.render_notebook()
```
