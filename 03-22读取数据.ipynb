## 任务一：文章分类

```python
keywords = [
"旅游",     "活动",     "节庆",     "特产",   "交通",
"酒店",     "景区",     "景点",     "文创",   "文化",
"乡村旅游", "民宿",     "假日",     "假期",   "游客",
"采摘",     "赏花",     "春游",     "踏青",   "康养",
"公园",     "滨海游",   "度假",     "农家乐", "剧本杀",
"旅行",     "徒步",     "工业旅游", "线路",   "自驾游",
"团队游",   "攻略",     "游记",     "包车",   "玻璃栈道",
"游艇",     "高尔夫",   "温泉"
]
```

```python
# 停用词
def load_stopwords(filename):
    with open(filename, encoding='utf-8') as f:
        words = [word.strip() for word in f]
    return words
```

```python
stopwords = load_stopwords("cn_stopwords2.txt")
```

```python
seg_list = [word for word in jieba.cut(string) if word not in stopwords]
wc = Counter(seg_list)
wc.keys() & set(keywords)
```

### 批量分析

```python
inter = []
wcounter = []
for row in weixin.itertuples():
    string = row.content
    seg_list = [word for word in jieba.cut(string.replace("\n", "")) if word not in stopwords]
    wc = Counter(seg_list)
    wcounter.append(wc.most_common())
    inter.append(set(wc.keys()) & set(keywords))
```

```python
weixin['wc'] = wcounter
weixin['inter'] = inter
```

```python
weixin['title_len'] = weixin.title.apply(len)
weixin['content_len'] = weixin.content.apply(len)
```

```python
weixin.sort_values(by="content_len", ascending=False)
```

```python
weixin.title_len.hist()
```

```python
weixin.content_len.hist()
```

```python
# 不含关键字的内容
inter_df = weixin[weixin.inter != set()]
```

```python
inter_df.sort_values(by="inter", key=lambda x: x.apply(len), ascending=False)
```

```python
# 打标签
weixin['label'] = 0
vec = weixin.inter.apply(lambda x: len(x))
weixin['label'][vec > 1] = 1
```
